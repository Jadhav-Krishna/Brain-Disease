```python
import numpy as np
import pandas as pd
from scipy.signal import resample
import matplotlib.pyplot as plt

# Step 1: Load the CSV file
file_path = r'C:\Users\ASUS\OneDrive\Desktop\Brain Disease\Epileptic Seizure\Epileptic Seizure Recognition.csv'  # Your CSV file path
data = pd.read_csv(file_path)

# Debug: Print column names
print("Column Names in CSV:", data.columns)

# Step 2: Extract EEG data
# Assuming EEG data is in columns 'X1' to 'X178' (based on the column names)
eeg_columns = [col for col in data.columns if col.startswith('X')]  # Select columns starting with 'X'
data_points = data[eeg_columns].values.flatten()  # Combine all EEG data into a single array

# Step 3: Uniform segmentation
num_channels = 14
points_per_channel = len(data_points) // num_channels
eeg_channels = [data_points[i * points_per_channel:(i + 1) * points_per_channel] for i in range(num_channels)]

# Optional: Resample if necessary (e.g., reduce to 256 data points per channel)
resampled_channels = [resample(channel, 256) for channel in eeg_channels]

# Step 4: Create a DataFrame for the 14 signals
resampled_df = pd.DataFrame(resampled_channels).transpose()  # Transpose to have signals as columns
resampled_df.columns = [f'Channel_{i + 1}' for i in range(num_channels)]  # Name the columns

# Step 5: Save the DataFrame to a new CSV file
output_path = r'C:\Users\ASUS\OneDrive\Desktop\Brain Disease\Epileptic Seizure\Processed_14_Signals.csv'
resampled_df.to_csv(output_path, index=False)
print(f"Processed EEG signals saved to: {output_path}")

# Optional: Visualize one of the channels
plt.plot(resampled_channels[0])  # Plot the first channel
plt.title("EEG Signal - Channel 1")
plt.xlabel("Time (resampled points)")
plt.ylabel("Amplitude")
plt.show()
```
